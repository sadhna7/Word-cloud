# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import numpy as np
```

# Generating word frequency

```python
from wordcloud import WordCloud ,  STOPWORDS
```

```python
mj_novel = open(r"C:\Users\sadhana\mj.txt", "r", encoding="utf-8").read()
def freq(text):
    word_list=text.split()
    word_freq=[]
    for word in word_list:
        if word not in word_freq:
            word_freq.append(word)
    for i in range(0,len(word_freq)):
        print(word_freq[i],word_list.count(word_freq[i]))
freq(mj_novel)
```

```python
stopwords = set(STOPWORDS)
```

```python
mj_wc = WordCloud(background_color = "white",max_words = 2000,stopwords = stopwords)
mj_wc.generate(mj_novel)
```

```python
plt.figure(figsize=(14,18))
plt.imshow(mj_wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```

# Superimposing wordcloud on mask

```python
mj_mask = np.array(Image.open(r"C:\Users\sadhana\Downloads\mike1.jpg"))
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(mj_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
mj_wc= WordCloud(background_color="white",max_words=2000,mask=mj_mask,stopwords=stopwords)
mj_wc.generate(mj_novel)
fig=plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(mj_wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```
